```python
from io import StringIO

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import cmocean
```

# Modified cmocean.balance
Use pure white at center

```python
cmocean_balance = np.genfromtxt("https://raw.githubusercontent.com/matplotlib/cmocean/refs/heads/main/cmocean/rgb/balance-rgb.txt")
```

```python
width = 0.05  # transition over 5% of colormap
idx_start = int(cmocean_balance.shape[0] * (0.5 - width))
idx_mid = int(cmocean_balance.shape[0] * 0.5)
idx_end = int(cmocean_balance.shape[0] * (0.5 + width))
cmocean_balance_mod = cmocean_balance.copy()
for i in range(3):
    cmocean_balance_mod[idx_start:idx_mid, i] = np.linspace(cmocean_balance[idx_start, i], 1, idx_mid - idx_start)
    cmocean_balance_mod[idx_mid:idx_end, i] = np.linspace(1, cmocean_balance[idx_end, i], idx_end - idx_mid)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(6, 2), dpi=300, tight_layout=True, facecolor="black")

axs[0].imshow(np.linspace(0, 1, 256)[np.newaxis, :], aspect='auto', cmap="cmo.balance")
axs[0].set_axis_off()
axs[0].set_title("Original from cmocean", color="white")

axs[1].imshow(np.linspace(0, 1, 256)[np.newaxis, :], aspect='auto', cmap=mpl.colors.LinearSegmentedColormap.from_list("", cmocean_balance_mod))
axs[1].set_axis_off()
axs[1].set_title("Modified with white center", color="white")

plt.show()
```

# Modified cmocean.amp

```python
cmocean_amp = np.genfromtxt("https://raw.githubusercontent.com/matplotlib/cmocean/refs/heads/main/cmocean/rgb/amp-rgb.txt")
```
